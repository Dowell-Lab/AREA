```python
def make_gene_and_comorbids_df(list_of_comorbids, agene_ensmbl):
  merge_df = gene_exp_df[[agene, "Patient"]]
  for acomorbid in list_of_comorbids:
    onegenedf = comorbid_df[[acomorbid, "Patient"]]
    merge_df = onegenedf.merge(merge_df, on="Patient")
    print ("There are ", merge_df[acomorbid].sum(), "people with ",acomorbid)
  print ("The mean expression of gene ",agene, " is ", merge_df[agene].mean())
  return merge_df
```

```python
def cutoff_probablity(cuttoff, raw_data_T21, comorbid, agene_ensmbl, agene):
  cutoff_df = raw_data_T21.copy()
  cutoff_df["above_cuttoff"] = cutoff_df[agene_ensmbl]>cuttoff
  # Split the dataframe based on the 'above_cuttoff' column
  cutoff_df_true = cutoff_df[cutoff_df['above_cuttoff'] == True]
  cutoff_df_false = cutoff_df[cutoff_df['above_cuttoff'] == False]
  # Calculate the count of 1s and 0s in the 'obstructive_sleep_apnea_syndrome' column for each group
  count_true = cutoff_df_true['obstructive_sleep_apnea_syndrome'].value_counts().to_frame()
  count_false = cutoff_df_false['obstructive_sleep_apnea_syndrome'].value_counts().to_frame()
  count_true['percentage'] = count_true['count'] / count_true['count'].sum()
  percent_with_comorbid_above_cutoff = count_true[count_true.index==1]["percentage"].to_list()[0]
  people_above_cuttoff = count_true['count'].sum()
  count_false['percentage'] = count_false['count'] / count_false['count'].sum()
  percent_with_comorbid_below_cutoff = count_false[count_false.index==1]["percentage"].to_list()[0]
  people_below_cuttoff = count_false['count'].sum()
  return people_above_cuttoff,  people_below_cuttoff, percent_with_comorbid_above_cutoff, percent_with_comorbid_below_cutoff

```

```python
# prompt: I would like to run cutoff_probablity with a list of cuttoffs starting at 450 and running to 900 every 50

cutoffs = list(range(425, 901, 20))
comorbid = 'obstructive_sleep_apnea_syndrome' # Example comorbid, replace as needed

results = []
for cutoff in cutoffs:
  people_above_cuttoff, people_below_cuttoff, percent_with_comorbid_above_cutoff, percent_without_comorbid_above_cutoff = cutoff_probablity(cutoff, raw_data_T21, comorbid, agene_ensmbl, agene)
  results.append([cutoff, people_above_cuttoff, people_below_cuttoff, percent_with_comorbid_above_cutoff, percent_without_comorbid_above_cutoff])

results_df = pd.DataFrame(results, columns=['Cutoff', 'People Above Cutoff', 'People Below Cutoff', 'Percent with Comorbid Above Cutoff', 'Percent without Comorbid Above Cutoff'])
results_df
```
