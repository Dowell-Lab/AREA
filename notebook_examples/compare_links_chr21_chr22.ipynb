# Find pairs predicted by PSEA

```python
chr21_df_nona_threshold = chr21df_nona[chr21df_nona["p_value_BenjaminiHochberg"]<0.1]
chr22_df_nona_threshold = chr22df_nona[chr22df_nona["p_value_BenjaminiHochberg"]<0.1]
```

```python
chr21_df_nona_threshold = chr21_df_nona_threshold.merge(genedf, how ="inner", on="value")
chr22_df_nona_threshold = chr22_df_nona_threshold.merge(genedf, how ="inner", on="value")
```

```python
chr22_df_nona_threshold["NES"] = chr22_df_nona_threshold["NES"].astype(float)
chr21_df_nona_threshold["NES"] = chr21_df_nona_threshold["NES"].astype(float)
```

```python
chr22_df_nona_threshold
```

# How many binary_attribute do we pair

```python
chr21_genes_per_disorder = chr21_df_nona_threshold["binary_attribute"].value_counts().reset_index()
chr21_genes_per_disorder.columns = ["binary_attribute", "comorbidity linked to x chr21 genes"]
chr22_genes_per_disorder = chr22_df_nona_threshold["binary_attribute"].value_counts().reset_index()
chr22_genes_per_disorder.columns = ["binary_attribute", "comorbidity linked to x chr22 genes"]
genes_per_disorder = chr21_genes_per_disorder.merge(chr22_genes_per_disorder, how ="outer", on="binary_attribute")
genes_per_disorder = genes_per_disorder.fillna(0) # Replace NaN values with 0
genes_per_disorder.sort_values(["comorbidity linked to x chr21 genes"])
```

```python
chr21_df_nona_threshold_direction = chr21_df_nona_threshold[chr21_df_nona_threshold["NES"]>0]
chr22_df_nona_threshold_direction = chr22_df_nona_threshold[chr22_df_nona_threshold["NES"]>0]
chr21_genes_per_disorder_direction = chr21_df_nona_threshold_direction["binary_attribute"].value_counts().reset_index()
chr21_genes_per_disorder_direction.columns = ["binary_attribute", "comorbidity linked to x chr21 genes NES pos"]
chr22_genes_per_disorder_direction = chr22_df_nona_threshold_direction["binary_attribute"].value_counts().reset_index()
chr22_genes_per_disorder_direction.columns = ["binary_attribute", "comorbidity linked to x chr22 genes NES pos"]
genes_per_disorder = genes_per_disorder.merge(chr21_genes_per_disorder_direction, how ="outer", on="binary_attribute")
genes_per_disorder = genes_per_disorder.merge(chr22_genes_per_disorder_direction, how ="outer", on="binary_attribute")
```

```python
genes_per_disorder = genes_per_disorder.fillna(0) # Replace NaN values with 0
genes_per_disorder.sort_values(["comorbidity linked to x chr21 genes"])
```
