# Find pairs predicted by PSEA

```python
chr21_df_nona_threshold = chr21df_nona[chr21df_nona["p_value_BenjaminiHochberg"]<0.1]
chrother_df_nona_threshold = chrotherdf_nona[chrotherdf_nona["p_value_BenjaminiHochberg"]<0.1]
```

```python
chr21_df_nona_threshold = chr21_df_nona_threshold.merge(genedf, how ="inner", on="value")
chrother_df_nona_threshold = chrother_df_nona_threshold.merge(genedf, how ="inner", on="value")
```

```python
chrother_df_nona_threshold["NES"] = chrother_df_nona_threshold["NES"].astype(float)
chr21_df_nona_threshold["NES"] = chr21_df_nona_threshold["NES"].astype(float)
```

```python
chr21_df_nona_threshold
```

```python
chrother_df_nona_threshold
```

# How many binary_attribute do we pair

```python
chr21_genes_per_disorder = chr21_df_nona_threshold["binary_attribute"].value_counts().reset_index()
chr21_genes_per_disorder.columns = ["binary_attribute", "comorbidity linked to x chr21 genes"]
chrother_genes_per_disorder = chrother_df_nona_threshold["binary_attribute"].value_counts().reset_index()
chrother_genes_per_disorder.columns = ["binary_attribute", "comorbidity linked to x "+otherchr+" genes"]
genes_per_disorder = chr21_genes_per_disorder.merge(chrother_genes_per_disorder, how ="outer", on="binary_attribute")
genes_per_disorder = genes_per_disorder.fillna(0) # Replace NaN values with 0
genes_per_disorder.sort_values(["comorbidity linked to x chr21 genes"])
```

```python
chr21_df_nona_threshold_direction = chr21_df_nona_threshold[chr21_df_nona_threshold["NES"]>0]
chrother_df_nona_threshold_direction = chrother_df_nona_threshold[chrother_df_nona_threshold["NES"]>0]
chr21_genes_per_disorder_direction = chr21_df_nona_threshold_direction["binary_attribute"].value_counts().reset_index()
chr21_genes_per_disorder_direction.columns = ["binary_attribute", "comorbidity linked to x chr21 genes NES pos"]
chrother_genes_per_disorder_direction = chrother_df_nona_threshold_direction["binary_attribute"].value_counts().reset_index()
chrother_genes_per_disorder_direction.columns = ["binary_attribute", "comorbidity linked to x "+otherchr+" genes NES pos"]
genes_per_disorder = genes_per_disorder.merge(chr21_genes_per_disorder_direction, how ="outer", on="binary_attribute")
genes_per_disorder = genes_per_disorder.merge(chrother_genes_per_disorder_direction, how ="outer", on="binary_attribute")
```

```python
chr21_df_nona_threshold_direction = chr21_df_nona_threshold[chr21_df_nona_threshold["NES"]<0]
chrother_df_nona_threshold_direction = chrother_df_nona_threshold[chrother_df_nona_threshold["NES"]<0]
chr21_genes_per_disorder_direction = chr21_df_nona_threshold_direction["binary_attribute"].value_counts().reset_index()
chr21_genes_per_disorder_direction.columns = ["binary_attribute", "comorbidity linked to x chr21 genes NES neg"]
chrother_genes_per_disorder_direction = chrother_df_nona_threshold_direction["binary_attribute"].value_counts().reset_index()
chrother_genes_per_disorder_direction.columns = ["binary_attribute", "comorbidity linked to x "+otherchr+" genes NES neg"]
genes_per_disorder = genes_per_disorder.merge(chr21_genes_per_disorder_direction, how ="outer", on="binary_attribute")
genes_per_disorder = genes_per_disorder.merge(chrother_genes_per_disorder_direction, how ="outer", on="binary_attribute")
```

```python
genes_per_disorder = genes_per_disorder.fillna(0) # Replace NaN values with 0
genes_per_disorder.sort_values(["comorbidity linked to x chr21 genes"])
```
