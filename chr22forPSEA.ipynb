```python
genedf = pd.read_csv("/Users/maryallen/psea/testdata/genes.csv")
genedf
```

```python
chronlyfiles = ["21", "22", "13", "18"]
for chronly in chronlyfiles:
    genedf_chr22 = genedf[genedf["seqnames"]==chronly]
    genedf_chr22 = genedf_chr22[genedf_chr22["gene_id"].isin(countdf.columns)]
    keepnames = ["Patient"]+genedf_chr22["gene_id"].to_list()
    countdf_22 = countdf[keepnames]
    countdf_22.to_csv("/Users/maryallen/psea/testdata/value_expression_chr"+chronly+".csv")
```
